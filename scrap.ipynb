```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
driver = webdriver.Chrome("C:/scrapping/chromedriver.exe")
```

```python
driver.get("http://rongdhonustudio.com/DhakaBus.html")
```

```python
content = driver.page_source
soup = BeautifulSoup(content)
```

```python
info = soup.find('div',attrs={'class' : 'accordion'})
info_list = info.find(id= 'myUL')
li = info_list.findAll('li')
# children = li.findChildren("a" , recursive=False)
# name = info.findall('btn btn-primary')
dhaka_bus_info=[]
names=[]
routes=[]
all_routes={}

for inf in li:
    route=[]
    name=inf.find('a',attrs={'class','btn btn-primary'})
    names.append(name.text)
    card=inf.find('div',attrs={'class','collapse'})
    card_body=card.find('div',attrs={'class','card card-body'})
    card_body=card_body.text.replace('\n','').replace('Route: ','')
    info_body=" ".join(card_body.split())
    route=info_body.split(" - ")
    routes.append(route)
#     all_routes={'name':name.text ,'route':routes}
#     dhaka_bus_info.append(all_routes)

all_routes={'name':names,'route':routes}
df=pd.DataFrame.from_dict(all_routes)
df.to_csv(index=False)
# print(all_routes)
                       
```
